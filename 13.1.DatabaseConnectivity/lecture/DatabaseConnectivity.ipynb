## Примеры

К некоторым СУБД можно подключиться как по ODBC, так и по JDBC.

Образец строчки подключения JDBC

`jdbc:sqlserver://[serverName[\instanceName][:portNumber]][;property=value[;property=value]]`

Образец строчки подключения ODBC

`Driver={SQL Server};Server=[serverName];Database=[instanceName][;property=value]`

Шаблон для кастомного (не JDBC или ODBC) PostgreSQL драйвера psycopg2

`host=[serverName] port=[portNumber] dbname=[instanceName] user=[username] password=[password]`

Шаблон для libpq

`postgresql://[user[:password]@][netloc][:port][/dbname][?param1=value1&...]`

## Курсор (cursor)

Цитаты Великого PEP'а

These objects represent a database cursor, which is used to manage the context of a fetch operation. Cursors created from the same connection are not isolated, i.e., any changes done to the database by a cursor are immediately visible by the other cursors. Cursors created from different connections can or can not be isolated, depending on how the transaction support is implemented.

````python
class TrackRepository:
    def __init__(self, connection):
        self.connection = connection

    def getAll(self) -> tp.Sequence[tp.Tuple[tp.Any, ...]]:
        cursor = self.connection.cursor()
        cursor.execute("SELECT * FROM tracks")
        return cursor.fetchall()

    def findByName(self, name: str) -> tp.Sequence[tp.Tuple[tp.Any, ...]]:
        cursor = self.connection.cursor()
        cursor.execute("SELECT * FROM tracks WHERE Name LIKE :name", {
            'name': f'%{name}%',
        })
        return cursor.fetchall()

    def findByComposer(self, composer: str) -> tp.Sequence[tp.Tuple[tp.Any, ...]]:
        cursor = self.connection.cursor()
        cursor.execute("SELECT * FROM tracks WHERE Composer LIKE :composer", {
            'composer': f'%{composer}%',
        })
        return cursor.fetchall()

    def findByNameAndComposer(self, name: str, composer: str) -> tp.Sequence[tp.Tuple[tp.Any, ...]]:
        cursor = self.connection.cursor()
        cursor.execute("SELECT * FROM tracks WHERE Name LIKE :name AND Composer LIKE :composer", {
            'name': f'%{name}%',
            'composer': f'%{composer}%',
        })
        return cursor.fetchall()


connection = sqlite3.connect('chinook.db')
track_repo = TrackRepository(connection)
track_repo.findByNameAndComposer(name='name', composer='composer')
````

### Table, Field

````python
from pypika import Table, Field

tracks = Table('tracks')
q = Query.from_(tracks) \
    .where(Field('Name').like('%Music%')).select('TrackId', 'Name') \
    .where(tracks.TrackId < 100)

print(str(q))
````

````python
# экранирование произошло автоматически
q = Query.from_(tracks) \
    .where(Field('Name').like('\' OR 1=1')).where(tracks.TrackId < 100) \
    .select('TrackId', 'Name')
print(str(q))
````

### Parameters

````python
from pypika import Parameter

# найдите отличие в выводе двух принтов :)
q = Query.from_(tracks) \
    .where(Field('Name').like(':name')).where(tracks.TrackId < 100) \
    .select('TrackId', 'Name')
print(str(q))

q = Query.from_(tracks) \
    .where(Field('Name').like(Parameter(':name'))).where(tracks.TrackId < 100) \
    .select('TrackId', 'Name')
print(str(q))
````

### Joins

````python
# порядок задания фильтраций/джойнов/select не важен. Все ради этого!
q = Query.from_(tracks) \
    .where(Field('Name').like(Parameter(':name'))).where(tracks.TrackId < 100) \
    .select('TrackId', 'Name')

albums = Table('albums')

# использование .on и .using одновременно отваливается с неочевидной ошибкой, к сожалению
q = q.inner_join(albums)
# q = q.using('AlbumId')
q = q.on(tracks.AlbumId == albums.AlbumId)
q = q.select(albums.Title)
str(q)
````

### Упражнение

Вывести топ плейлистов, по убыванию стоимостей. Вместе с плейлистами вывести их ранк. Цену плейлиста определить как сумму цен треков.

````python
import sqlite3
import typing as tp

connection = sqlite3.connect('chinook.db')

def execute_and_get_dicts(query: 'QueryBuilder', query_args: tp.Mapping[str, tp.Any] | None = None) -> tp.Sequence[dict[str, tp.Any]]:
    cursor = connection.cursor()
    cursor.execute(str(query), query_args or {})
    column_names = [full_column_info[0] for full_column_info in cursor.description]
    tuple_results = cursor.fetchall()
    return [{name: value for name, value in zip(column_names, tup)} for tup in tuple_results]
````

````python
from pypika import Query, analytics as an

tracks = Table('tracks')
playlist_track = Table('playlist_track')
playlists = Table('playlists')

playlist_cost = fn.Sum(tracks.UnitPrice).as_('cost')
rank = an.Rank().orderby(playlist_cost, order=Order.desc).as_('rank')

ranked_playlists = Query.from_(playlists) \
    .select(
        rank,
        tracks.name,
        playlist_cost,
    ) \
    .inner_join(playlist_track).using('PlayListId') \
    .inner_join(tracks).using('TrackId') \
    .groupby(tracks.name) \
    .orderby(rank)

query = Query.from_(ranked_playlists) \
    .select(
        ranked_playlists.rank,
        ranked_playlists.name,
        ranked_playlists.cost,
    ) \
    .where(ranked_playlists.rank < 10)

query
````

# Related models

````python
import json
from pypika.terms import CustomFunction

class PlaylistRepository:
    def __init__(self, connection):
        self.connection = connection

    # {"id": 1, "name": "Playlist 1", "tracks": []}

    def get(self, id: int) -> tp.Iterable[tp.Mapping[str, tp.Any]] | None:
        query = Query.from_(playlists) \
            .select(playlists.PlaylistId, playlists.Name, tracks.Name.as_("TrackName")) \
            .inner_join(playlist_track).using('PlaylistId') \
            .inner_join(tracks).using('TrackId') \
            .where(playlists.PlaylistId == Parameter(":id"))
        return execute_and_get_dicts(query, {'id': id})

    def get_transformed_by_python(self, id: int) -> tp.Mapping[str, tp.Any] | None:
        cursor = self.connection.cursor()
        query = Query.from_(playlists) \
            .select(playlists.PlaylistId, playlists.Name, tracks.Name.as_("TrackName")) \
            .inner_join(playlist_track).using('PlaylistId') \
            .inner_join(tracks).using('TrackId') \
            .where(playlists.PlaylistId == Parameter(":id"))
        cursor.execute(str(query), {'id': id})
        result = cursor.fetchall()
        related_tracks = []
        for row in result:
            related_tracks.append({"name": row[2]})

        if len(related_tracks) == 0:
            return None
        else:
            return {
                "id": result[0][0],
                "name": result[0][1],
                "tracks": related_tracks,
            }

    def get_by_two_queries(self, id: int):
        playlist_query = Query.from_(playlists) \
            .select(playlists.PlaylistId, playlists.Name) \
            .where(playlists.PlaylistId == Parameter(":id"))

        tracks_query = Query.from_(tracks) \
            .select(tracks.Name) \
            .inner_join(playlist_track).using('TrackId') \
            .where(playlist_track.PlaylistId == Parameter(":id"))

        playlist = execute_and_get_dicts(playlist_query, {'id': id})[0]
        related_tracks = execute_and_get_dicts(tracks_query, {'id': id})
        playlist["tracks"] = related_tracks
        return playlist

    def get_by_json_build_obj(self, id: int):
        JSON_BUILD_OBJECT = CustomFunction("json_object", ['key', 'value'])
        JSON_AGG = CustomFunction("json_group_array", ['arg'])

        track_json = JSON_BUILD_OBJECT('Name', tracks.Name)
        track_agg_json = JSON_AGG(track_json)

        query = Query.from_(playlists) \
            .select(playlists.PlaylistId, playlists.Name, track_agg_json.as_('tracks')) \
            .inner_join(playlist_track).using('PlaylistId') \
            .inner_join(tracks).using('TrackId') \
            .where(playlists.PlaylistId == Parameter(":id")) \
            .groupby(playlists.PlaylistId, playlists.Name)

        result = execute_and_get_dicts(query, {'id': id})[0]
        result["tracks"] = json.loads(result["tracks"])
        return result

connection = sqlite3.connect('chinook.db')
playlist_repo = PlaylistRepository(connection)
playlist_repo.get(id=1)
````

# Object-Relational Mapping (ORM)

````python
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.orm import declarative_base

# единственный СУБД зависимый объект
engine = create_engine('sqlite+pysqlite:///chinook.db', echo=True)

Base = declarative_base()  # класс-родитель для всех классов-моделей (таблиц)
````

### Модели

````python
class SimpleTrack(Base):
    __tablename__ = 'tracks'  # атрибут, содержащий имя моделируемой таблицы

    # список атрибутов
    track_id = Column(Integer, name='TrackId', primary_key=True)    # хотя бы один атрибут
                                                                    # должен являться первичным ключом
    Name = Column(String)
    GenreId = Column(Integer)
    UnitPrice = Column(Float)
    MediaTypeId = Column(Integer)
    Milliseconds = Column(Integer)

# класс-модель во многом дублирует структуру БД, sqlalchemy умеет создавать таблицы по моделям
````

### Сессия

- контейнер для объектов-моделей
- сессия содержит в себе объект подключение к базе, которое открывается “лениво”
- все взаимодействия с базой проходят через сессию
- объекты добавляются в сессию через `.add()` или при поиске объектов через эту сессию
- запросы в рамках сессии также выполняются “лениво” и обернуты в транзакцию
- чем-то похоже на подключение, но с дополнительными плюшками
- объекты в рамках сессии уникализируются по первичному ключу


````python
from sqlalchemy.orm import sessionmaker

Session = sessionmaker(bind=engine, future=True, expire_on_commit=False)  # фабрика сессий

session = Session()  # конкретные сессии
session2 = Session()
````

````python
from sqlalchemy import select

# пока что просто создали питонячий объект, в БД изменений нет
some_track = SimpleTrack(MediaTypeId=1, Name='new track', GenreId=1, UnitPrice=1.0, Milliseconds=297823)

# все еще ничего не произошло. Но теперь этот объект привязан к сессии, дальше станет понятно что это значит
session.add(some_track)

# пробуем достать только что созданный объект: произошло добавление строки и тут же обратный выбор
res = session.execute(select(SimpleTrack).filter_by(Name='new track')).scalar()
````

### Связи между таблицами (relations)

````python
import sqlalchemy as sa
from sqlalchemy.orm import relationship


playlist_track = sa.Table(
    "playlist_track",
    Base.metadata,
    Column("PlaylistId", Integer, sa.ForeignKey("playlists.PlaylistId"), primary_key=True),
    Column("TrackId", Integer, sa.ForeignKey("tracks.TrackId"), primary_key=True),
)

class Track(Base):
    __tablename__ = 'tracks'
    __table_args__ = {'extend_existing': True}  # чтобы не было конфликта с ранее объявленной моделью Track

    track_id = Column(Integer, name='TrackId', primary_key=True)
    Name = Column(String)
    GenreId = Column(Integer)
    UnitPrice = Column(Float)
    MediaTypeId = Column(Integer)
    Milliseconds = Column(Integer)

    playlists = relationship(
        "Playlist",
        secondary=playlist_track,
        back_populates="tracks"
    )


class Playlist(Base):
    __tablename__ = 'playlists'
    __table_args__ = {'extend_existing': True}

    playlist_id = Column(Integer, name='PlaylistId', primary_key=True)
    Name = Column(String)

    tracks = relationship(
        "Track",
        secondary=playlist_track,
        back_populates="playlists"
    )
````

# Выводы

### Raw SQL
Легко и быстро писать небольшие SQL-выражения.
Из минусов - тяжело параметризировать запросы, СУБД-зависимая имплементация

### ORM
Удобно для быстрого прототипирования, когда много однотипных CRUD endpoint-ов.
Минусы:
 - Неявная генерация SQL, проблемы с производительностью, усложняет отладку и профилирование
 - Ленивые поля
 - Утечка абстракции
 - Сложные запросы пишутся очень сложно
 - Тяжеловесные сущности, занимают много памяти, хранятся в identity map
 - Высокий порог входа

### SQL Builder
Оптимальное решение, решает проблемы сырого sql не добавляя лишней магии.

Иногда в приложениях используются комбинированные подходы, для применения лучшего из подходов по месту.
